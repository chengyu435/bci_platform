```python
# -*- coding: utf-8 -*-
"""
测试 EDF 数据读取模块

测试文件：
    PhysioNet EEG Motor Movement/Imagery Dataset
    S001R01.edf

测试内容：
    1. 文件头读取
    2. 通道信息读取
    3. 采样率读取
    4. 数据矩阵读取
    5. 简单数据检查
    6. 简单频谱测试准备
"""

import os
import sys
import numpy as np

# 添加父目录到sys.path，使得能导入bci_platform包
# 如果 __file__ 存在则使用，否则使用当前工作目录
if '__file__' in globals():
    sys.path.insert(0, os.path.dirname(os.path.dirname(os.path.abspath(__file__))))
else:
    sys.path.insert(0, os.path.dirname(os.getcwd()))

from bci_platform.io.edf_bdf_reader import EDFBDFReader







def main():
    file_path = os.path.join("data", "S001R01.edf")

    if not os.path.exists(file_path):
        raise FileNotFoundError(
            f"未找到测试文件：{file_path}\n"
            f"请确认 S001R01.edf 已放入 data 文件夹。"
        )

    print("开始读取 EDF 文件:")
    print(file_path)

    reader = EDFBDFReader(file_path)

    eeg = reader.read(preload=True, physical=True)

    reader.print_summary()




    print("\n========== 测试完成 ==========")


if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt

fs = eeg.main_sampling_rate

ch_idx = 0
start_time = 0
duration = 5

start_sample = int(start_time * fs)
end_sample = int((start_time + duration) * fs)

x = eeg.data[ch_idx, start_sample:end_sample]
t = np.arange(start_sample, end_sample) / fs

plt.figure(figsize=(12, 4))
plt.plot(t, x, linewidth=0.8)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (uV)")
plt.title(f"Channel {eeg.channel_names[ch_idx]}")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
